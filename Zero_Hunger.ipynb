```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
path = 'FAO E.xlsx'
plt.figure(figsize=(20,20))
#Can write the path directly as well
df = pd.read_excel(path)
data_item=df['Item'].head(20)
data_Y2004=df['Y2008'].head(20)
plt.barh(data_item,data_Y2004)
first_50 = df.head(50)
first_50.sort_values(by=['Y2008'], ascending=True)
print(first_50[['Item','Unit','Y2008']])
print("\t\t")
min=first_50[['Item','Y2008']].min()
print("min",min)
max=first_50[['Item','Y2008']].max()
print("max:",max)
print("\t\t")
print("\t\t")
print(first_50[['Item','Y2008']].describe())
print("\t")
print('Mean',df['Y2008'].mean())
print('Median',df['Y2008'].median())
print('Mode',df['Y2008'].mode())
print("\t\t")
sns.catplot(x='Y2008',kind='box',data=df)
print('Mean',df['Y2008'].mean())
print( 'Median' ,df['Y2008'].median())
print( 'Mode' ,df[ 'Y2008' ].mode())

plt.xlabel("item")
plt.ylabel("data_Y2004")
plt.show()
```

```python
Q1=df['Y2010'].quantile(.25)
Q3=df['Y2010'].quantile(.75)
IQR=Q3-Q1
print('IQR',Q3-Q1)
lb=Q1-1.5*IQR

ub=Q3+1.5*IQR
print(ub)

dfout=df[(df['Y2010']>lb) & (df['Y2010']<ub) ]
dfout.shape
dfout.tail()
sns.catplot(x='Y2010',kind='box',data=dfout)
```

```python
from seaborn.distributions import stats
import numpy as np
zscore=stats.zscore(df['Y2010'])
print('Zscore',zscore)

filter=((zscore > -1) &(zscore < 2))
newdf=df[filter]
newdf.tail()
sns.catplot(x='Y2010',kind='box',data=newdf)
```

```python
#can be done by using either train_size or test_size
from sklearn.model_selection import train_test_split
x = df['Y2013']
y = df['Y2012']
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size = 0.75, shuffle = True)
from sklearn.linear_model import LinearRegression
print(x_train.shape)
print(x_test.shape)

```

```python
import numpy as np
x_train=np.array(x_train).reshape(-1,1)
y_train=np.array(y_train).reshape(-1,1)

model=LinearRegression()
model. fit(x_train,y_train)

print (model. coef_)
print(model.intercept_)
```

```python
y_test=np.array(y_test).reshape(-1,1)
y_pred=model.predict(y_test)
print(y_pred)
```

```python
#Performance analysis

from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
import math

print('Mean absolute error:', mean_absolute_error(y_test, y_pred))

print("Mean squared error:", mean_squared_error(y_test, y_pred))
print('variance score:',r2_score(y_test, y_pred))

print('Root Mean Squared Error:',math.sqrt(mean_squared_error(y_test, y_pred)))
```
